```python
# Split the data into train, validate, and test
train, test = train_test_split(df, train_size=0.8, random_state=1234)
train, validate = train_test_split(train, train_size=0.7, random_state=1234)
train.head()
```

3. Work through all of the steps outlined in the lesson, from setting the baseline to selected a model and evaluating the final model on your test data.

```python
X_train = train.drop(columns = ['angle'])
y_train = train.angle

X_validate = validate.drop(columns=['angle'])
y_validate = validate.angle

X_test = test.drop(columns=['angle'])
y_test = test.angle
```

```python
plt.hist(y_train, color = "skyblue", ec="black")
plt.xlabel("Angles")
plt.ylabel("Number of Angles")
plt.show()
```

```python
y_train = pd.DataFrame(y_train)
    # turn it into a single pandas dataframe
y_validate = pd.DataFrame(y_validate)
    # wrap them as dataframes

# 1. Predict angle_pred_mean
    # 2 different aselines of mean and medium
angle_pred_mean = y_train['angle'].mean()
y_train['angle_pred_mean'] = angle_pred_mean
y_validate['angle_pred_mean'] = angle_pred_mean

# 2. compute angle_pred_median
    # same process as mean (above)
angle_pred_median = y_train['angle'].median()
y_train['angle_pred_median'] = angle_pred_median
y_validate['angle_pred_median'] = angle_pred_median

# 3. RMSE of angle_pred_mean
rmse_train = mean_squared_error(y_train.angle, y_train.angle_pred_mean)**1/2
    # stick with root mean square error
        # not your only option but that is what we will be using here
            # just because it is eaiest to us and explain
    # remember when you call you it will be your y_true and y_pred
rmse_validate = mean_squared_error(y_validate.angle, y_validate.angle_pred_mean)**1/2
    # do the same thing for the validate set as done above for the train set
    
print("RMSE using Mean\nTrain/In-Sample: ", round(rmse_train, 2), 
      "\nValidate/Out-of-Sample: ", round(rmse_validate, 2))

# 4. RMSE of angle_pred_median
rmse_train = mean_squared_error(y_train.angle, y_train.angle_pred_median)**1/2

rmse_validate = mean_squared_error(y_validate.angle, y_validate.angle_pred_median)**1/2

print("RMSE using Median\nTrain/In-Sample: ", round(rmse_train, 2), 
      "\nValidate/Out-of-Sample: ", round(rmse_validate, 2))

# median comes out a little higher than mean 
    # but a little lower than our validation set
```

## Linear Regression (OLS)

```python
# create the model object
lm = LinearRegression(normalize=True)

# fit the model to our training data. We must specify the column in y_train, 
# since we have converted it to a dataframe from a series! 
lm.fit(X_train, y_train.angle)
    # just call y_train.actual_target

# predict train
y_train['angle_pred_lm'] = lm.predict(X_train)

# evaluate: rmse
rmse_train = mean_squared_error(y_train.angle, y_train.angle_pred_lm)**1/2

# predict validate
y_validate['angle_pred_lm'] = lm.predict(X_validate)

# evaluate: rmse
rmse_validate = mean_squared_error(y_validate.angle, y_validate.angle_pred_lm)**1/2
    # make sure you are using x_validate an not x_train

print("RMSE for OLS using LinearRegression\nTraining/In-Sample: ", rmse_train, 
      "\nValidation/Out-of-Sample: ", rmse_validate)
```

## LassoLars

```python
# create the model object
lars = LassoLars(alpha=1.0)

# since we have converted it to a dataframe from a series! 
lars.fit(X_train, y_train.angle)
    # fit the thing

# predict train
y_train['angle_pred_lars'] = lars.predict(X_train)

# evaluate: rmse
rmse_train = mean_squared_error(y_train.angle, y_train.angle_pred_lars)**1/2

# predict validate
y_validate['angle_pred_lars'] = lars.predict(X_validate)

# evaluate: rmse
rmse_validate = mean_squared_error(y_validate.angle, y_validate.angle_pred_lars)**1/2

print("RMSE for Lasso + Lars\nTraining/In-Sample: ", rmse_train, 
      "\nValidation/Out-of-Sample: ", rmse_validate)
```

## Tweedie Regressor (GLM)

```python
# create the model object
glm = TweedieRegressor(power=1, alpha=0)

# fit the model to our training data. We must specify the column in y_train, 
# since we have converted it to a dataframe from a series! 
glm.fit(X_train, y_train.angle)

# predict train
y_train['angle_pred_glm'] = glm.predict(X_train)

# evaluate: rmse
rmse_train = mean_squared_error(y_train.angle, y_train.angle_pred_glm)**1/2

# predict validate
y_validate['angle_pred_glm'] = glm.predict(X_validate)

# evaluate: rmse
rmse_validate = mean_squared_error(y_validate.angle, y_validate.angle_pred_glm)**1/2

print("RMSE for GLM using Tweedie, power=1 & alpha=0\nTraining/In-Sample: ", rmse_train, 
      "\nValidation/Out-of-Sample: ", rmse_validate)
```

## Polynomial Features

```python
# make the polynomial features to get a new set of features
pf = PolynomialFeatures(degree=2)

# fit and transform X_train_scaled
X_train_degree2 = pf.fit_transform(X_train)

# transform X_validate_scaled & X_test_scaled
X_validate_degree2 = pf.transform(X_validate)
    # dont call fit transform just call transform
        # we only fit train not validate
X_test_degree2 = pf.transform(X_test)
```
