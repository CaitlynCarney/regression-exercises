1. Write a function named plot_variable_pairs that accepts a dataframe as input and plots all of the pairwise relationships along with the regression line for each pair.

```python
def plot_variable_pairs():
    '''accepts a dataframe as input
    split into train, validate, and test
    plots all of the pairwise relationships along with the regression line for each pair.'''
    # Take in the dataframe
    df = acquire.acquire_all_telco()
    # split the data
    train, test = train_test_split(df, test_size=.2, random_state=1234)
    train, validate = train_test_split(train, test_size=.3, random_state=1234)
    # plot pairwise relationships using train data
    sns.pairplot(train, hue="churn")
    return train
```

2. Write a function named months_to_years that accepts your telco churn dataframe and returns a dataframe with a new feature tenure_years, in complete years as a customer.

```python
def months_to_years():
    '''accepts your telco churn dataframe 
    splits data into train, validate, and test
    returns a dataframe with a new feature tenure_years, in complete years as a customer.'''
    # accept data
    df = acquire.acquire_all_telco()
    # split data
    train, test = train_test_split(df, test_size=.2, random_state=1234)
    train, validate = train_test_split(train, test_size=.3, random_state=1234)
    # make tenure_year
    train['tenure_years'] = round(df.tenure / 12, 2)
    # rename tenure to tenure_month
    train = train.rename(columns={'tenure': 'tenure_month'})
    return train
```

3. Write a function named plot_categorical_and_continuous_vars that accepts your dataframe and the name of the columns that hold the continuous and categorical features and outputs 3 different plots for visualizing a categorical variable and a continuous variable.



```python
def plot_categorical_and_continuous_vars():
    '''accepts your dataframe
    accepts the name of the columns that hold the continuous and categorical features
    outputs 3 different plots for visualizing a categorical variable and a continuous variable.
    '''
    # get it
    df = acquire.acquire_all_telco()
    # split it
    train, test = train_test_split(df, test_size=.2, random_state=1234)
    train, validate = train_test_split(train, test_size=.3, random_state=1234)
    # make the visuals
    plt.subplots(3, 1, figsize=(13,25), sharey=True)
    sns.set(style="darkgrid")
        # Monthly Charges and Churn
    plt.subplot(3,1,1)
    sns.barplot(x='churn',y='monthly_charges',data=train)
    plt.title('Distribution of Monthly Charges')
        # Tenure and Churn
    plt.subplot(3,1,2)
    sns.boxplot(x='churn',y='tenure',data=train,palette='rainbow')
    plt.title('Distribution of Tenure')
        # Total Charges and Churn
    plt.subplot(3,1,3)
    sns.stripplot(x="total_charges", y="churn", data=train, palette='rainbow')
    plt.title('Distribution of Total Charges')
    return train
```
